```python
#data training
data1 = data.head(len(data)-1)
```

```python
data1
```

```python
#storing independent and dependent data
days = list()
Adj_closes = list()
```

```python
#get the date and the open price values
data1_days = data1.loc[:,'Date']
data1_adj_close = data1.loc[:, 'Adj Close']
```

```python
#creating independent data set
for day in data1_days:
    #print(day)
    #print(day.split('-')[2])
    days.append( [int(day.split('-')[2])])
#creating the dependent datset
for adj_close in data1_adj_close:
    Adj_closes.append(float(adj_close))
```

```python
#checking the values stored
days, Adj_closes
```

```python
from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
regr = make_pipeline(StandardScaler(), SVR(C=1000.0,kernel = 'rbf', epsilon=0.2, gamma = 0.15))
regr.fit(days, Adj_closes)
```

```python
regr1 = make_pipeline(StandardScaler(), SVR(C=1000.0,kernel = 'linear', epsilon=0.2))
regr1.fit(days, Adj_closes)
```

```python
regr2 = make_pipeline(StandardScaler(), SVR(C=1000.0,kernel = 'poly', epsilon=0.2, degree = 2))
regr2.fit(days, Adj_closes)
```

```python
day = [[30]]
```

```python
pred = regr.predict(day)
pred
```

```python
regr1.predict(day)
```

```python
regr2.predict(day)
```

```python
#visualizing the results to check the bet fit kernel
import matplotlib.pyplot as plt

plt.figure(figsize =(15, 7))
plt.scatter(days, Adj_closes, color= 'black', label= 'Orginal Data')
plt.plot(days, regr.predict(days), color = 'orange', label ='RBF')
plt.plot(days, regr1.predict(days), color = 'red', label ='Poly')
plt.plot(days, regr2.predict(days), color = 'green', label='Linear')
plt.legend()
plt.show()
```
